```python
from elasticsearch import Elasticsearch
import json

server_file = "server_credentials.json"
```

```python
class QueryData:
    def __init__(self, server_file, extra_pars = {}):
        with open(server_file, 'r') as f:
            pars = json.load(f)
        pars.update(extra_pars)
        self.client = Elasticsearch(**pars)
```

```python
queryData = QueryData(server_file)
```

```python
queryData.client.indices.get_alias(index="*")
```
