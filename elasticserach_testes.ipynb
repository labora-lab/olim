```python
from elasticsearch import Elasticsearch
import json
from pprint import pprint

server_file = "server_credentials.json"
```

```python
class QueryData:
    def __init__(self, server_file, extra_pars = {}):
        with open(server_file, 'r') as f:
            pars = json.load(f)
        pars.update(extra_pars)
        self.pars = pars
        self.client = Elasticsearch(**pars)
```

```python
queryData = QueryData(server_file)
```

## Acessa um paciente específico

```python
query={
    "match": {"cd_usu_cadsus": 1347568}
}
data = queryData.client.search(index='cpf_prontuarios', query=query, scroll='5m', size=10)

print("Número de retornos: {}".format(len(data['hits']['hits'])))

data['hits']['hits'][0]
```

## Acessa alista de todos os pacientes

```python
aggs={
        "cadsus": {
            "terms": { "field": "cd_usu_cadsus" }
        }
    }
data = queryData.client.search(index='cpf_prontuarios', aggs=aggs, scroll='5m', size=10_000)

print("Número de retornos: {}".format(len(data['hits']['hits'])))

data['hits']['hits'][0]
```
